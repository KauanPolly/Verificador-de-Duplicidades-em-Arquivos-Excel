```python
import pandas as pd

def verificar_duplicidade_login_data(df):
    # Agrupar pelo login e verificar se há datas de vencimento duplicadas dentro do mesmo login
    for login, group in df.groupby('Login'):
        # Verificar duplicidade nas datas de vencimento para um mesmo login
        if group.duplicated(subset=['datavenc'], keep=False).any():
            # Se houver duplicata, pegar as datas duplicadas
            datas_duplicadas = group[group.duplicated(subset=['datavenc'], keep=False)]['datavenc'].tolist()
            print(f"Login {login} possui duplicidade nas datas de vencimento: {datas_duplicadas}")

# Exemplo de uso com o DataFrame 'titulos'
verificar_duplicidade_login_data(titulos)
```
